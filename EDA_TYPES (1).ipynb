# Bi-Varient Analysis


```python
sns.barplot(x='sex',y='expenses',data=df)
```

```python
sns.barplot(x='age',y='region',data=df)
```

```python
sns.jointplot(data=df,x='expenses',y='sex')
```

```python
sns.barplot(x='age',y='expenses',data=df)
```

```python
sns.jointplot(data=df,x='age',y='expenses',hue="sex")
```

```python
sns.boxplot(x='smoker',y='expenses',data=df,palette='rainbow')
```

```python
sns.violinplot(x="sex",y="expenses",data=df,palette='rainbow')
```

In case we have large datasets with 30-70+ features (variables), there might not be sufficient time to run each pair of varience through bivarient analysis one by one.
