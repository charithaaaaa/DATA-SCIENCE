Python Program to perform data Visualization using matplotlib and seaborn library

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv("auto-mpg.csv")
df

```

```python
iris=sns.load_dataset('iris') 
iris
```

Distribution plot represents how data is distributed at certain intervals. The intervals called as bins.
distplot
 display the distribution plot using taking number of cylinders on X-axis


```python
sns.displot(data= df, x='cylinders',bins=5) # bins as 5 intervals 
plt.xlabel("No. of Cylinders") 
plt.show()
```

```python
## kde display (kde-kernel density levels) 
sns.displot(data=df,x='cylinders',bins=5,kde=True)  
plt.xlabel("No. of Cylinders")
plt.show()
```

```python
 
#observe that maximum is given by cars having 4 cylinders
```

```python
## bivariate count plot
plt.text(5, 60, 'count plot',fontsize = 22)
sns.countplot(data=df,x='cylinders') 
plt.xlabel("No. of Cylinders") 
plt.ylabel("No. of cars")

```

```python
df['model-year'].unique()
```

```python
sns.countplot(data=df,x='cylinders',hue='model-year')
## hue is understand what  different origins are present in df
```

```python
## scatter plot 
sns.scatterplot(x='horsepower',y='acceleration',data=df)
plt.figure(figsize=(10,6))
plt.figure().set_figwidth(60)
plt.figure().set_figheight(2)
plt.rcParams['figure.figsize'] = [30, 10]


```

```python
sns.jointplot(x='horsepower',y='acceleration',data=df)
```

```python
sns.lineplot(x='horsepower',y='acceleration',data=df)
```

```python
sns.lmplot(x='cylinders',y='mpg', data=df)
```

```python
#heat map
iris1=iris.iloc[:10,0:4]
iris1
```

```python
sns.heatmap(iris1,annot=True)
```

```python
#pairplot respresents pairwise relationships in a dataset
sns.pairplot(iris,hue='species')
```

```python
# subplots
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.scatterplot(y='acceleration',x='horsepower',hue='model-year',data=df) 
plt.subplot(2,2,2)
sns.scatterplot(y='acceleration',x='horsepower',data=df) 
plt.subplot(2,2,3)
sns.boxplot(y='acceleration',x='model-year',data=df)
```
